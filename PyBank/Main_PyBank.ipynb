```python
import pandas as pd
import csv

path = 'Resources/budget_data.csv'
df = pd.read_csv(path)
df
```

```python
df['Shifted value']=df['Profit/Losses'].shift(1)
df['Change']=df['Profit/Losses']-df['Shifted value']
df
```

```python
df1=df.sort_values('Change')
df1
```

```python
date_greatest_decrease = df1.loc[44,'Date']
```

```python
df2=df.sort_values('Change',ascending=False)
df2
```

```python
date_greatest_increase = df2.loc[25,'Date']
```

```python
months = df.shape[0]
print('Total Months: '+str(months))

total = df['Profit/Losses'].sum()
print('Total: '+str(total))

Average_change = df['Change'].mean()
print('Average change: $'+str(Average_change))

greatest_increase = df['Change'].max()
print(f'Greatest increase in Profits: {date_greatest_increase} {greatest_increase}')

greatest_decrease = df['Change'].min()
print(f'Greatest decrease in Profits: {date_greatest_decrease} {greatest_decrease}')
```
